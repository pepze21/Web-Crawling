```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

driver = webdriver.Chrome('C:/users/2020/chromedriver/chromedriver.exe')

driver.get('https://youtube-rank.com/board/bbs/board.php?bo_table=youtube')

html = driver.page_source

soup = BeautifulSoup(html, 'lxml')

rank = soup.select('td.rank')
channel = soup.select('td.subject > a')
subscribers = soup.select('td.subscriber_cnt')
views = soup.select('td.view_cnt')
videos = soup.select('td.video_cnt')
hits = soup.select('td.hit > strong')

data_list = []

for i in range(len(rank)):
    data_list.append([rank[i].text, channel[i].text, subscribers[i].text, \
                      views[i].text, videos[i].text, hits[i].text])
columns = ['순위', '채널', '구독자', '조회수', '영상 수', 'HIT']
df = pd.DataFrame(data_list, columns = columns)

# 공백 제거
for column in df.columns.to_list():
    df[column] = df[column].str.strip()
df.to_excel('C:/users/2020/Youtube_ranking_crawling&visualization/Youtube_ranking_crawling&visualization.xlsx', index=False)
df = df.set_index('순위')
df
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

driver = webdriver.Chrome('C:/users/2020/chromedriver/chromedriver.exe')

driver.get('https://youtube-rank.com/board/bbs/board.php?bo_table=youtube')

html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')


channels = soup.select('tr.aos-init')



columns = ['Rank', 'Category', 'Name', 'Subscribers', 'Views', 'Videos', 'Hits']

data = []

for channel in channels:
    rank = channel.select('td.rank')[0].text.strip()
    category = channel.select('p.category')[0].text.strip()
    name = channel.select('td.subject h1 a')[0].text.strip()
    subscribers = channel.select('td.subscriber_cnt')[0].text.strip()
    views = channel.select('td.view_cnt')[0].text.strip()
    videos = channel.select('td.video_cnt')[0].text.strip()
    hits = channel.select('td.hit')[0].text.replace('\n', '')[:-3]
    data.append([rank, category, name, subscribers, views, videos, hits])

df = pd.DataFrame(data, columns = columns)
df = df.set_index('Rank')
df
```
